# Sorting

```python
import numpy as np
import pandas as pd
```

```python
# read csv
df = pd.read_csv('ShanHaiJing_raw.csv',encoding='gbk')
df_sorted = pd.DataFrame()
df

```

```python
# label set
labels = set(df['物种'])
print(labels)
print(df.shape)

# sort with labels
for label in labels:
  dataItem = pd.DataFrame(df[df['物种']==label])
  df_sorted = df_sorted.append(dataItem)
  print(dataItem.shape)
print(df_sorted.shape)
```

```python
df_sorted
```

# Save to File

```python
output = os.path.join(path, 'ShanHaiJing_sorted.csv')
df_sorted.to_csv(output,encoding = 'utf-8')
```
